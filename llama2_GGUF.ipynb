```python
from huggingface_hub import HfApi
api = HfApi()

# gguf 파일 경로
file_path = "llama-2-ko-7b-jiyeon.gguf"

# Hugging Face 저장소 이름
repo_name = "eomjiyeon/customLLM"

# 파일 업로드
api.upload_file(
    path_or_fileobj=file_path,
    path_in_repo="llama-2-ko-7b-custom.gguf",  # 저장소 내 파일 경로
    repo_id=repo_name,
    repo_type="model",  # 모델 저장소에 업로드
)
```
